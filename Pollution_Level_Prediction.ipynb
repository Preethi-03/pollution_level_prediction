AQI from Original Data for CO(in mg/m³), NO2(in µg/m³),NOx as NO2 (approximate)

```python
def calculate_co_aqi(concentration):
    breakpoints = [
        (0.0, 1.0, 0, 50),
        (1.1, 2.0, 51, 100),
        (2.1, 10.0, 101, 200),
        (10.1, 17.0, 201, 300),
        (17.1, 34.0, 301, 400),
        (34.1, 50.0, 401, 500),
    ]
    for bp_lo, bp_hi, i_lo, i_hi in breakpoints:
        if bp_lo <= concentration <= bp_hi:
            return round((i_hi - i_lo) / (bp_hi - bp_lo) * (concentration - bp_lo) + i_lo)
    return None
def calculate_no2_aqi(concentration):
    breakpoints = [
        (0, 40, 0, 50),
        (41, 80, 51, 100),
        (81, 180, 101, 200),
        (181, 280, 201, 300),
        (281, 400, 301, 400),
        (401, 1000, 401, 500),
    ]
    for bp_lo, bp_hi, i_lo, i_hi in breakpoints:
        if bp_lo <= concentration <= bp_hi:
            return round((i_hi - i_lo) / (bp_hi - bp_lo) * (concentration - bp_lo) + i_lo)
    return None
def calculate_nox_aqi(concentration):
    return calculate_no2_aqi(concentration)

# Apply functions on your DataFrame
df['CO_AQI'] = df['CO(GT)'].apply(calculate_co_aqi)
df['NO2_AQI'] = df['NO2(GT)'].apply(calculate_no2_aqi)
df['NOx_AQI'] = df['NOx(GT)'].apply(calculate_nox_aqi)

# Calculate overall AQI (maximum of individual AQIs)
df['Overall_AQI'] = df[['CO_AQI', 'NO2_AQI', 'NOx_AQI']].max(axis=1)
print(df[['CO(GT)', 'NO2(GT)', 'NOx(GT)', 'CO_AQI', 'NO2_AQI', 'NOx_AQI', 'Overall_AQI']].head())
```

```python
def classify_aqi(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Satisfactory'
    elif aqi <= 200:
        return 'Moderate'
    elif aqi <= 300:
        return 'Poor'
    elif aqi <= 400:
        return 'Very Poor'
    elif aqi <= 500:
        return 'Severe'
    else:
        return 'Out of Range'

df['AQI_Category'] = df['Overall_AQI'].apply(classify_aqi)
df[['Overall_AQI', 'AQI_Category']].head(10)

```

SVM

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Define your AQI columns as features
features = ['CO_AQI', 'NO2_AQI', 'NOx_AQI']
X = df[features]

# Define the target column
y = df['AQI_Category']

# Drop any rows with missing values
df_clean = df.dropna(subset=features + ['AQI_Category'])
X = df_clean[features]
y = df_clean['AQI_Category']

# Encode AQI category labels (Good, Moderate, etc.) to numbers
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train the SVM model
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_model.fit(X_train, y_train)

# Predict and Evaluate
y_pred = svm_model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
# Convert predictions back to category labels
y_pred_labels = le.inverse_transform(y_pred)

# Create a new column in df_clean for predicted AQI category (only for test data)
df_clean.loc[X_test.index, 'Predicted_AQI_Category'] = y_pred_labels

```

```python
df_clean = df_clean.reset_index(drop=True)
X_train, X_test, y_train, y_test = train_test_split(
    df_clean[features], y_encoded, test_size=0.2, random_state=42
)

# Train and predict
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
y_pred_labels = le.inverse_transform(y_pred)

# Add predictions correctly after resetting
df_clean.loc[X_test.index, 'Predicted_AQI_Category'] = y_pred_labels

```

```python
def compute_co_aqi(co):
    if 0 <= co <= 1:
        return (50/1.0) * co
    elif 1 < co <= 2:
        return ((100-51)/(2.0-1.1)) * (co - 1.1) + 51
    elif 2 < co <= 10:
        return ((200-101)/(10-2.1)) * (co - 2.1) + 101
    elif 10 < co <= 17:
        return ((300-201)/(17-10.1)) * (co - 10.1) + 201
    elif 17 < co <= 34:
        return ((400-301)/(34-17.1)) * (co - 17.1) + 301
    elif 34 < co <= 50:
        return ((500-401)/(50-34.1)) * (co - 34.1) + 401
    else:
        return None
```
